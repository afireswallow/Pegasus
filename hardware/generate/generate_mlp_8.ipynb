```python
def generate_p4():
    tables = [
        ("ip_len_total_len", ["hdr.feature.ip_len", "hdr.feature.ip_total_len"],                "set_ip_len_total_len"          ),
        ("protocol_tos",      ["hdr.feature.protocol", "hdr.feature.tos"],                        "set_protocol_tos"              ),
        ("ttl_offset",      ["hdr.feature.ttl", "hdr.feature.offset"],                            "set_ttl_offset"                ),
        ("max_min_byte",      ["hdr.feature.max_byte", "hdr.feature.min_byte"],         "set_max_min_byte"            ),
        ("max_min_ipd",     ["hdr.feature.max_ipd", "hdr.feature.min_ipd"],          "set_max_min_ipd"            ),
        ("ipd",             ["hdr.feature.ipd"],       "set_ipd"          ),
    ]

    actions = [
        ("set_ip_len_total_len"    , "ig_md.linear1", True ),
        ("set_protocol_tos"        , "ig_md.linear3", True ),
        ("set_ttl_offset"          , "ig_md.linear1", False),
        ("set_max_min_byte"        , "ig_md.linear3", False),
        ("set_max_min_ipd"         , "ig_md.linear1", False),
        ("set_ipd"                 , "ig_md.linear3", False),
    ]
    p4_code = ""
    for name, prefix0, first in actions:
        p4_code += generate_action_4(name, prefix0, first)
    stage = 0
    table_num = 0
    for name, key, action in tables:
        p4_code += generate_table(name, key, action, stage + table_num // 2)
        table_num += 1
    return p4_code

with open("generated_code.p4", "w") as f:
    f.write(generate_p4())
```
