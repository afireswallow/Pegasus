```python
def generate_action_cnn(name, lists, nums):
    cnn_outputs = []
    for num in range(len(nums)):
        for i in range(len(lists)):
            for j in range(lists[i][0], lists[i][1] + 1):
                cnn_outputs.append(f"ig_md.cnn_{nums[num]}_{i+1}.l16_{j}")

    ins = ""
    inputs = []
    outputs = ''
    for i in range(len(cnn_outputs)):
        inputs.append(f"in{i}")
        if i != 0:
            ins = ins + ', '
        ins = ins + f'bit<16> in{i}'
        outputs += f'\t\t{cnn_outputs[i]} = in{i} + {cnn_outputs[i]};\n'

    return f"""
    action {name}({ins}) {{
{outputs}
    }}
    """
```
